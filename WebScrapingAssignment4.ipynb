```python
page
```

```python
soup= BeautifulSoup(page.content)
soup
```

```python
header_tag1
```

```python
df = pd.DataFrame(header_tag1, columns=["Header"])
print(df)
```

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get(" https://www.imdb.com/list/ls523837749/?ref_=tt_sw_pks_dec23_i")
```

```python
search= driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[2]/div/div[1]/div/div[4]/div[3]/div[2]/div[2]/h3/a")
```

```python
series_title=[]
title_tags=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[2]/div/div[1]/div/div[4]/div[3]/div[2]/div[2]/h3')
for i in title_tags:
    title=i.text
    series_title.append(title)
series_title
```

```python
df=pd.DataFrame({'Title':series_title})
df
```
